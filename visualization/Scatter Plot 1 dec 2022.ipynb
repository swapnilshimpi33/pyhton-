# Subplots

```python
#import necessary libraries
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

%matplotlib inline
```

```python
a = np.linspace(1,15,40)
b = np.random.random_sample(40)
```

```python
# create anempty canvas
```

```python
fig, ax = plt.subplots(2,2)

ax[0,0].plot(a,b)
ax[0,1].scatter(a,b)
ax[1,0].boxplot(a)
ax[1,1].plot(b,a)
```

fig : create as acanvas for plotting


ax  : decides axis for plots

```python
fig, ax = plt.subplots(2,2)
plt.suptitle('2 * 2 plots')
```

```python
df =pd.read_csv(r'C:\Users\asus\Downloads\Datasets-main\Cars.csv')
```
