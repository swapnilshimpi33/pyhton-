```python
glaxo = glaxo[['Date','Close']]
beml = beml[['Date','Close']]
```

```python
#data description
```

```python
#change the datatype of to Datetime
```

```python
glaxo['Date'] = pd.DatetimeIndex(glaxo['Date'])
```

```python
beml['Date'] = pd.DatetimeIndex(beml['Date'])
```

```python
glaxo = glaxo.set_index(glaxo['Date'])
```

```python
glaxo
```

```python
beml = beml.set_index(beml['Date'])
```

```python
beml
```

```python
# inplace=True used to drop/delete column permanently
glaxo.drop(columns=['Date'],inplace=True)
```

```python
glaxo
```

```python
# inplace=True used to drop/delete column permanently
beml.drop(columns=['Date'],inplace=True)
```

```python
beml.columns
```

```python
beml
```

```python
#plot closing value for both glaxo and beml
sns.set_style('darkgrid')
plt.rcParams['figure.figsize']=(10,5)
plt.rcParams['figure.dpi'] = 200
plt.plot(glaxo['Close'])
plt.xlabel('Time')
plt.ylabel('Close')
```

```python
plt.plot(beml['Close'])
plt.xlabel('Time')
plt.ylabel('Close')
```

```python
#calculate gains
```

```python
glaxo['gain'] = glaxo ['Close'].pct_change(periods=1)
beml['gain'] = beml ['Close'].pct_change(periods=1)
```

```python
glaxo
```

```python
beml
```

```python
#drop the null record
```

```python
glaxo.dropna(inplace=True)
beml.dropna(inplace=True)
```

```python
glaxo
```

```python
beml
```

```python
# plot gain values
```

```python
plt.plot(glaxo['gain'])
plt.xlabel('Time')
plt.ylabel('Gain')
```

```python
plt.plot(beml['gain'])
plt.axhline(beml['gain'].mean(),color='y')
plt.xlabel('Time')
plt.ylabel('Gain')
```

```python
#subplot for distribution of BEML and GLAXO
```

```python
plt.subplot(1,2,1)
sns.distplot(glaxo['gain'],color='r',label='GLAXO')
plt.xlabel('Time')
plt.ylabel('Gain')
plt.legend()
plt.title('Distribution Graph')

plt.subplot(1,2,2)
sns.distplot(beml['gain'],color='b',label= 'BEML')
plt.legend()
plt.title('Distribution Graph')
```
